```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings;
warnings.filterwarnings('ignore')
# %matplotlib notebook
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
sns.violinplot(cars.displacement);
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',data=cars);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])])

plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
               inner='quartiles')

plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
               scale='count'
               )

plt.legend(loc=2);
```

```python
sns.violinplot(x='cylinders', y='displacement', hue='origin',
               data=cars[cars.origin.isin(['japan','europe'])],
               split=True,
               scale='count',
               scale_hue=False
               )

plt.legend(loc=2);
```

```python
(cars[cars.origin.isin(['japan','europe'])].groupby('cylinders')
 .origin 
 .value_counts()
 )
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement, order=['japan', 'europe','usa']);
```

```python
sns.violinplot(x=cars.origin, y=cars.displacement, linewidth=4)
```

```python
sns.violinplot(x=cars.cylinders, y=cars.displacement, bw=0.2)
```
