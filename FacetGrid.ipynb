```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings;
warnings.filterwarnings('ignore')
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.set(style='darkgrid', font_scale=1.5)
```

```python
g = sns.PairGrid(iris)
g.map(sns.scatterplot);
```

```python
type(g)
```

```python
g = sns.PairGrid(iris)
g.map(sns.histplot)
```

```python
g = sns.PairGrid(iris)
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.histplot);
```

```python
g = sns.PairGrid(iris, corner=True)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
g.map_diag(sns.boxplot);
```

```python
g = sns.PairGrid(iris, hue='species', diag_sharey=False, vars=['petal_length', 'petal_width'])
g.map_offdiag(sns.scatterplot)
g.map_diag(sns.kdeplot);
```

```python
g = sns.PairGrid(iris,
                 x_vars=['sepal_length', 'sepal_width', 'petal_width'],
                 y_vars=['petal_length'])
g.map(sns.scatterplot);
```

```python
g = sns.PairGrid(iris, hue='species', diag_sharey=False, height=2, aspect=2, palette='viridis')
g.map_offdiag(sns.scatterplot, s=80, alpha=.8, ec='black')
g.map_diag(sns.kdeplot, fill=True, lw=3);
```

```python
g = sns.PairGrid(iris, hue='species', diag_sharey=False)
g.map_offdiag(sns.scatterplot, s=80, alpha=.7, ec='black')
g.map_diag(sns.kdeplot, fill=True, lw=3);
```
