```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.set(style='darkgrid', font_scale=1.5)
```

```python
sns.pairplot(tips);
```

```python
tips.dtypes
```

```python
sns.pairplot(tips, diag_kind='kde');
```

```python
sns.pairplot(tips, diag_kind='kde', kind='reg');
```

```python
g = sns.pairplot(tips)
```

```python
g = sns.pairplot(tips, diag_kind='kde', kind='reg')
g.map_upper(sns.kdeplot, n_levels=6, color='xkcd:wine red');
```

```python
# New column 'weekend'
tips['weekend'] = tips.day.isin(['Sat', 'Sun'])
```

```python
tips.head()
```

```python
sns.pairplot(tips, hue='weekend');
```

```python
tips.dtypes
```

```python
sns.pairplot(tips, hue='weekend', vars=['total_bill', 'tip']);
```

```python
sns.pairplot(tips, hue='weekend', x_vars=['total_bill', 'size'],y_vars=['tip'], kind='reg',height=4)
```

```python
del tips['weekend']
```

```python
from ipywidgets import interactive 

def size_widget(height=2.5, aspect=1):
    sns.pairplot(tips, height=height, aspect=aspect)
```

```python
interactive(size_widget, height=(1, 3.5, 0.5), aspect=(0.5, 2, 0.25))
```
