```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings;
warnings.filterwarnings('ignore')
```

```python
cars = sns.load_dataset('mpg')
```

```python
cars.head()
```

```python
cars.groupby('origin').cylinders.value_counts()
```

```python
(cars.groupby('origin')
.cylinders
.value_counts()
.unstack())
```

```python
origin_cyl = (cars.groupby('origin')
.cylinders
.value_counts()
.unstack()
.fillna(0)
)
```

```python
sns.heatmap(origin_cyl);
```

```python
test_array = np.array([[1,2], [3,4]])
test_array
```

```python
sns.heatmap(test_array)
```

```python
origin_cyl.T
```

```python
sns.heatmap(origin_cyl.T);
```

```python
cars.corr()
```

```python
sns.heatmap(cars.corr())
```

```python
sns.heatmap(origin_cyl, cmap='Blues')
```

```python
plt.figure(figsize=(13,6))

sns.heatmap(cars.corr(), cmap='RdBu', annot=True,vmin=-1, vmax=1);
```

```python
plt.figure(figsize=(13,6))

sns.heatmap(cars.corr(), cmap='coolwarm', annot=True,vmin=-1, vmax=1);
```

```python
plt.figure(figsize=(13,6))

sns.heatmap(origin_cyl, cmap='coolwarm', annot=True, fmt=".2f",
            annot_kws={'fontsize':14, 'fontweight':'bold', 'fontfamily':'serif'});
```

```python
plt.figure(figsize=(13,6))

sns.heatmap(cars.corr(), cmap='coolwarm', linewidth=3);
```

```python
plt.figure(figsize=(13,6))

sns.heatmap(cars.corr(), cmap='coolwarm', linewidth=3, linecolor='black', annot=True);
```

```python
labels = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Model Year']
```

```python
plt.figure(figsize=(13,6))

plt.xticks(rotation=90)

sns.heatmap(cars.corr(), cmap='coolwarm', xticklabels=labels, yticklabels=labels);
```

```python
plt.figure(figsize=(13,6))

sns.heatmap(cars.corr(), cmap='coolwarm', square=True);
```
