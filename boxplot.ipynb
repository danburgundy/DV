```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
import warnings;
warnings.filterwarnings('ignore')
```

```python
cars = sns.load_dataset('mpg').dropna()

cars.shape
```

```python
cars.head()
```

```python
sns.set_style('whitegrid')
```

```python
cars.cylinders.value_counts()
```

```python
cars = cars[cars.cylinders.isin([4,6,8])]
```

```python
cars.cylinders.value_counts()
```

```python
sns.boxplot(cars.mpg);
```

```python
cars.mpg.describe()
```

```python
sns.boxplot(x=cars.origin, y=cars.mpg);
```

```python
sns.boxplot(x='origin', y='mpg', data=cars, hue='cylinders');
```

```python
cars.model_year.describe()
```

```python
cars['newer_model'] = cars.model_year > 76
```

```python
sns.boxplot(x='origin', y='mpg', hue='newer_model', data=cars);
```

```python
sns.boxplot(x='origin', y='mpg', order=['japan', 'europe', 'usa'], data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', hue='newer_model',order=['japan', 'europe', 'usa'], hue_order=[True,False],data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', hue='newer_model',color='g',data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', width=0.5,data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', linewidth=3.5,data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', whis=1,data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', fliersize=20,data=cars);
```

```python
sns.boxplot(x='mpg', y='origin', showcaps=False,data=cars);
```
