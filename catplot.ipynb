```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
#%matplotlib notebook
```

```python
# load car data and drop rows with missings
cars = sns.load_dataset('mpg').dropna()

# filter to 4,6,8 cylinders cars 
cars = cars[cars.cylinders.isin([4,6,8])]

# add old/new column
cars['type'] = ['old' if x <= 76 else 'new' for x in cars.model_year]
```

```python
cars.head()
```

```python
cars.shape
```

```python
sns.set_style('white')
```

```python
sns.catplot(x = 'horsepower', data=cars, kind='box');
```

```python
sns.catplot(x = 'horsepower', data=cars, kind='violin');
```

```python
sns.catplot(x='origin', y='horsepower', 
            hue='cylinders',
            data=cars, 
            kind='box');
```

```python
sns.catplot(x='origin', y='horsepower', 
            hue='cylinders',
            data=cars, 
            kind='bar');
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin');
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin');
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            row='type',
            row_order=['new', 'old']);
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin',
            palette='winter',
            whis=1
            );
```

```python
sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='strip',
            col='origin',
            s=10,
            alpha=0.5,
            linewidth=3
            );
```

```python
sns.set(style='darkgrid', font_scale=1.5)
g = sns.catplot(x='cylinders', y='horsepower', 
            data=cars, 
            kind='box',
            col='origin',
            );
```
